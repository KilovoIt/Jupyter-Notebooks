```python
import ipywidgets
from ipywidgets import widgets
from ipywidgets import GridspecLayout, Button, Layout
import numpy as np
from traitlets import traitlets 

def create_expanded_button(description, button_style):
    return Button(description=description, button_style=button_style, layout=Layout(height='30px', width='auto'))


grid = GridspecLayout(5, 4, layout=Layout(width='230px', height='170px'))

grid[1, 0] = create_expanded_button('1', '')
grid[1, 1] = create_expanded_button('2', '')
grid[1, 2] = create_expanded_button('3', '')
grid[2, 0] = create_expanded_button('4', '')
grid[2, 1] = create_expanded_button('5', '')
grid[2, 2] = create_expanded_button('6', '')
grid[3, 0] = create_expanded_button('7', '')
grid[3, 1] = create_expanded_button('8', '')
grid[3, 2] = create_expanded_button('9', '')
grid[4, 1] = create_expanded_button('0', '')
grid[4, 0] = create_expanded_button('.', '')
grid[4, 2] = create_expanded_button('exec', 'info')
grid[1, 3] = create_expanded_button('+', 'info')
grid[2, 3] = create_expanded_button('-', 'info')
grid[3, 3] = create_expanded_button('✕', 'info')
grid[4, 3] = create_expanded_button('÷', 'info')

class Calculator(widgets.Text):
    
    def __init__(self, ans=0, value=0, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.ans = ans 

    def cleaner(self):
        if '=' in self.value:
            self.value =''
    
    def addition(self):
        if '=' in self.value:
            self.value = str(self.ans)
        self.value+='+'
    
    def subtraction(self):
        if '=' in self.value:
            self.value = str(self.ans)
        self.value+='-'

    def division(self):
        if '=' in self.value:
            self.value = str(self.ans)
        self.value+='/'

    def multiplicaion(self):
        if '=' in self.value:
            self.value = str(self.ans)
        self.value+='*'


    def execution(self):
        self.ans = eval(self.value)
        self.value = self.value + f'={self.ans}'





grid[0, :] = Calculator(layout=Layout(height='10px', width='auto'))

grid[4, 2].style.button_color='#012a4a'
grid[1, 3].style.button_color='#014f86'
grid[2, 3].style.button_color='#014f86'
grid[3, 3].style.button_color='#014f86'
grid[4, 3].style.button_color='#014f86' 

grid[1, 0].style.button_color='#89c2d9'
grid[1, 1].style.button_color='#89c2d9'
grid[1, 2].style.button_color='#89c2d9'
grid[2, 0].style.button_color='#89c2d9'
grid[2, 1].style.button_color='#89c2d9'
grid[2, 2].style.button_color='#89c2d9'
grid[3, 1].style.button_color='#89c2d9'
grid[3, 2].style.button_color='#89c2d9'
grid[1, 2].style.button_color='#89c2d9'
grid[4, 0].style.button_color='#89c2d9'
grid[4, 1].style.button_color='#89c2d9'
grid[3, 0].style.button_color='#89c2d9'




def on_click(b):
    grid[0, :].cleaner()
    grid[0, :].value+=str(b.description)

def addition(b):
    grid[0, :].addition()

def subtraction(b):
    grid[0, :].subtraction()

def multiplication(b):
    grid[0, :].multiplicaion()

def division(b):
    grid[0, :].division()

def execution(b):
    grid[0, :].execution()

grid[1, 0].on_click(on_click)
grid[1, 0].on_click(on_click)
grid[1, 1].on_click(on_click)
grid[1, 2].on_click(on_click)
grid[2, 0].on_click(on_click)
grid[2, 1].on_click(on_click)
grid[2, 2].on_click(on_click)
grid[3, 0].on_click(on_click)
grid[3, 1].on_click(on_click)
grid[3, 2].on_click(on_click)
grid[4, 1].on_click(on_click)
grid[4, 0].on_click(on_click)

grid[1, 3].on_click(addition)
grid[4, 2].on_click(execution)
grid[3, 3].on_click(multiplication)
grid[2, 3].on_click(subtraction)
grid[4, 3].on_click(division)







grid
```
